```python
# Definição da função menu:
def menu():
    print('\n')
    print("(1) Filmes em Exibição")
    print("(2) Disponibilidade do Lugar")
    print("(3) Lugares Pós-Compra")
    print("(4) Informações das Salas")
    print("(5) Acrescentar Nova Sala")
    print('\n')
    return int(input("Escolhe as opções possíveis."))


cinema1 = [(5,25,[2,3,6,16,8],"Mulan"),(6,58,[2,1,4,6,8,9],"Avatar"),(3,69,[2,1,63,8,9],"Divertidamente"),(1,25,[23,21,24,8,19],"Interceptor")]

# Definição das funções:
#1 
def listar(cinema):
    print("Lista de Salas: ")
    for sala in cinema:
        numS, nlugares, vendidos, filme = sala
        print("Sala Número: ", numS)
        print("Filmes em exibição: ", filme) 
        print("-------------------")


#2 
def disponivel(cinema, filmeExb, lugar):
    dispo = False
    for sala in cinema:
        numS, nlugares, vendidos, filme = sala
        if filmeExb == filme and lugar <= nlugares:
           if lugar not in vendidos:
            dispo = True
    print("Disponibilidade do lugar:", '\n')
    if dispo == True:
        print(dispo, " -> O lugar", lugar, "está disponível.")
    if dispo == False:
        print(dispo, " -> O lugar", lugar, "não está disponível.")
    return dispo

#3 
def vendebilhete(cinema, filmeExb, lugar):
    if disponivel(cinema, filmeExb, lugar):
        for sala in cinema:
            numS, nlugares, vendidos, filme = sala
            if filmeExb == filme:
                vendidos.append(lugar)
        print("O lugar", lugar, "foi registado.")
        print("Atualização das Salas do Cinema", cinema)
    



#4 
def listardisponibilidades(cinema):
    for sala in cinema:
        numS, nlugares, vendidos, filme = sala
        for numOcup in vendidos:
            numOcup = len(vendidos)
            lugDispo = nlugares - numOcup
 
        print("Na sala", numS, "O filme que está a ser exibido é: ", filme)
        print("Lugares disponíveis na sala: ", lugDispo)


#5
def inserirSala(cinema, sala):
    P = 0
    for s in cinema:
        if s[0] == sala[0]:
            P = P + 1
    if P > 0:
        print("Esta sala já existe no cinema.")
    if P == 0:
        cinema.append(sala)
        print("Salas | Filmes em Exibição:")
        for s in cinema:
            numS, nlugares, vendidos, filme = s
            print("Sala", numS, "---->", filme)




# Criar uma lista para inserir os lugares ocupados.
def criarLista(N):
    lista1 = []
    i = 0
    while i < N:
        elemento = int(input("Insere os lugares que estão ocupados na sala."))
        lista1.append(elemento)
        i = i + 1
    return lista1



def exercicio():
    option = 1
    while option != 0:
        option = menu()
        if option == 1:
            listar(cinema1) 
        if option == 2:
            disponivel(cinema1, str(input("Qual o filme que procuras em exibição?")), int(input("Qual o lugar que procuras?")))
        if option == 3:
            cinema = vendebilhete(cinema1, str(input("Qual o filme que procuras?")), int(input("Qual o lugar que procuras?")))
        if option == 4:
            listardisponibilidades(cinema1)
        if option == 5:
            inserirSala(cinema1, (int(input("Insere o número da sala.")), int(input("Insere o número de lugares da sala.")), criarLista(int(input("Quantos lugares estão ocupados na sala?"))), str(input("Insere o filme exibido na sala."))))
    print("Saiu do programa.") 

exercicio()
```
