* Crie uma função que calcula a distribuição da doença por escalões etários. Considere os seguintes escalões: [30-34], [35-39], [40-44], ...

* Crie uma função que imprime na forma de uma tabela uma distribuição;

```python
global populacao
global classe
global ValoresClasse

populacao = []
classe = []
valoresClasse = []


def menu():
    print("\n"+"(1) Distribuição por Sexo")
    print("(2) Distribuição por Faixa Etária")
    print("(3) Distribuição por Níveis de Colesterol")
    print("(0) Sair do Programa", "\n")
    return int(input("Escolhe uma das opções possíveis."))


#def RecolhaInf(nomeficheiro): 
file = open("myheart.csv", "r")
listaDeDados = []
file.readline()
for linha in file:
    dados = linha.split(",")
    listaDeDados.append(dados)
for dados in listaDeDados:
    dados[-1] = dados[-1].replace("\n", "")
    pessoa = (dados[0], dados[1], dados[2], dados[3], dados[4], dados[5])
    populacao.append(pessoa)
file.close()


#Função que faz distribuição por sexo
def DistSexo(Lista):
    global valoresClasse
    global classe 
    
    classe = ["Masculino", "Feminino"]
    masculino = 0
    feminino = 0
    for pessoa in Lista:
        idade, sexo, tensão, colesterol, batimento, temDoenca = pessoa
        temDoenca = int(temDoenca)
        sexo = str(sexo)
        if temDoenca != 0:
            if sexo == "F":
                feminino = feminino + 1
            else:
                masculino = masculino + 1
        valoresClasse = [masculino, feminino]



#Função que faz distribuição por faixa etária
def DistIdade(Lista2):
    global valoresClasse
    global classe

    classe = []
    valoresClasse = []
    menor = 25
    pessoaComDoenca = 0
    while menor <= 73:
        for pessoa in Lista2:
            idade, sexo, tensão, colesterol, batimento, temDoenca = pessoa
            temDoenca = int(temDoenca)
            if temDoenca != 0:
                if int(idade) >= menor and int(idade) <= menor + 4:
                    pessoaComDoenca = pessoaComDoenca + 1
        valoresClasse.append(pessoaComDoenca)
        NovaClasse = ("[" + str(menor) + "," + str(menor + 4) + "]")
        menor = menor + 5
        classe.append(NovaClasse)
        pessoaComDoenca = 0



#Função que faz distribuição por faixa etária
def DistColesterol(Lista3):
    global valoresClasse
    global classe

    classe = []
    valoresClasse = []
    menor = 0
    pessoaComDoenca = 0
    while menor <= 89:
        for pessoa in Lista3:
            idade, sexo, tensão, colesterol, batimento, temDoenca = pessoa
            temDoenca = int(temDoenca)
            if temDoenca != 0:
                if int(idade) >= menor and int(idade) <= menor + 9:
                    pessoaComDoenca = pessoaComDoenca + 1
        valoresClasse.append(pessoaComDoenca)
        NovaClasse = ("[" + str(menor) + "," + str(menor + 9) + "]")
        menor = menor + 10
        classe.append(NovaClasse)
        pessoaComDoenca = 0

#Função que constrói as tabelas
def tabelas(classe, valoresClasse):
    i = 0
    while i < len(classe):
        print(classe[i], "|", valoresClasse[i])
        i = i + 1
    
#Função "mestre"
def exercicio():
    option = 1
    while option != 0:
        option = menu()
        if option == 1:
            DistSexo(populacao)
            print("Sexo", "    ", "|", "Nº Utentes com Doença")
            print("----------------------------------")
            tabelas(classe, valoresClasse)
        if option == 2:
            DistIdade(populacao)
            print ("Idade", " ", "|", "Nº Utentes com Doença")
            print("----------------------------------")
            tabelas(classe, valoresClasse)
        if option == 3:
            DistColesterol(populacao)
            print("Nível", " ", "|", "Nº Utentes com Doença")
            print("----------------------------------")
            tabelas(classe, valoresClasse)
    print("Saiu do programa.")


exercicio()
```
